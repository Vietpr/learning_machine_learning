## Lp trnh hm h(x)  d on output vi parameter c sn
#### Univariate problem
####  lp trnh c hm h(x) vi NumPy, u tin ta load ma trn X v vector Theta.
#### Trong vn  univariate ny chng ta s d on li nhun ca mt xe thc n t dn s ca thnh ph. Ct u trong file l dn s, ct th hai l li nhun.

```python
import numpy as np  
X = np.loadtxt('univariate.txt', delimiter = ',')
# import file theta
Theta = np.loadtxt('univariate_theta.txt')
```

```python
# Vi X, ta cn phi thm x0 lun bng 1  thc hin php nhn ma trn. x0 c thm vo ct u tin ca X.
#Lu ct y
y = np.copy(X[:,-1])
#Chuyn ct u (x1) sang ct th 2
X[:,1] = X[:,0]
#i ct u thnh s 1
X[:,0] = 1
```

```python
#Tnh li nhun (n v 10000$)
predict = X @ Theta
#Chuyn li nhun v n v $
predict = 10000 * predict
#in cp dn s-li nhun u tin (n v dn s: ngi)
print('%d ngi : %.2f$' %(X[0,1]*10000,predict[0]))
```

```python
#Lu kt qu
np.savetxt('predicted_value.txt',predict,fmt = '%.6f')
```

```python
#Use char 
import matplotlib.pyplot as plt
#Plot gi tr thc t (khng ly ct bias 1 u)
#X[:,1:] l x-axis ca biu , khng ly ct u; y l y-axis, rx l red x, plot d liu bng du x mu 
plt.plot(X[:,1:],y,'rx')
```

#### Multivariate problem
D on gi nh ($) t kch thc (feet-vung) v s phong ng.

```python
Data = np.loadtxt('multivariate.txt', delimiter = ',')
Data_theta = np.loadtxt('multivariate_theta.txt')
```

```python
#To ma trn X bng kch thc ca Data
X = np.zeros((np.size(Data,0),np.size(Data,1)))
#Thm ct u bng 1
X[:,0] = 1
#Thm cc ct x1 -> xn
n = np.size(Data,1) - 1
X[:,1:] = Data[:,0:n]
```

```python
#Tnh gi nh (n v $)
predict = X @ Data_theta 
predict = predict * 10000
#in b din tch-s phng-gi u tin
print('%.2f feet-vung, %d phng ng : %.1f$' %(X[0,1], X[0,2], predict[0]))

#Lu kt qu
np.savetxt('predicted_value.txt',predict,fmt = '%.2f')
```
